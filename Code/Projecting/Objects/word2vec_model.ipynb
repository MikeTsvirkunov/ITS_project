```python
import gensim.models
from json import load
```

```python
with open('../../Data/Datasets/data.json', 'r', encoding='utf-8') as js:
    discriptors = load(js)
```

```python
list_of_all_worlds = list()
for i in discriptors.values():
    for j in i:
        for z in i[j]:
            list_of_all_worlds.append(z)
```

```python
list_of_all_worlds
```

```python
model = gensim.models.Word2Vec(sentences=list_of_all_worlds, min_count=2, vector_size=100)
```

```python
model.save('../../PipeLines/Classifications/word2vec-descriptors.h5')
```
